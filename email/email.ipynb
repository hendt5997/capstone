```python
import pandas as pd
```

```python
df_email = pd.read_csv('./email.csv')
```

```python
df_email.columns
```

```python
df_email.loc[0]
```

```python
df_email['date'] = pd.to_datetime(df_email['date'])
```

```python
#num external emails
df_email['calendarDay'] = df_email['date'].dt.date
df_email['dateString'] = df_email['calendarDay'].apply(lambda x: x.strftime('%Y%m%d'))
df_emailToUser_id = pd.read_pickle('../ldap/emailToUser_id.pkl').rename(columns={'user_id':'fromUser_id','email':'from'})
df_email = pd.merge(df_email, df_emailToUser_id, how='left', on='from')
df_email['userDate'] = df_email.apply(lambda x: x['fromUser_id'] + '_' + x['dateString'], axis=1)
df_email
```

```python
df_emailCounts = df_email[~(df_email['to'].str.contains('dtaa.com'))].groupby('userDate').agg({'size':'sum','attachments':'sum','id':'count'}).rename(columns={'size':'sumExtEmailSize','attachments':'sumExtEmailAttachments','id':'numExtEmails'})
```

```python
df_emailCounts.to_pickle('emailCounts.pkl')
```
